```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("../data/raw/aapl.csv", index_col="Date", parse_dates=True)
df.head()
```

```python
df.head()
```

```python
df['daily_return'] = (df['Close'] - df['Close'].shift(1)) / df['Close'].shift(1)
df[['Close', 'daily_return']].head()
```

```python
df['ma_20'] = df['Close'].rolling(window=20).mean()
```

```python
df[['Close', 'ma_20']].head(25)
```

```python
df['volatility_20'] = df['daily_return'].rolling(window = 20).std()
```

```python
df[['daily_return', 'volatility_20']].head(25)
```

```python
df['momentum_10'] = (df['Close'] - df['Close'].shift(10)) / df['Close'].shift(10)
```

```python
df[['Close', 'momentum_10']].head(15)
```
